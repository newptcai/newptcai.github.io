## Harmonic numbers and the pandemic

In the last post we have seen how to approximate the $m$-th harmonic number, i.e.,
$$
H_m = \sum_{i=1}^m \frac{1}{i}
$$

This number appears in the famous [coupon collection problem](https://en.wikipedia.org/wiki/Coupon_collector%27s_problem). But let give you another place where it appears.

Assume that we live in a place called Path Town which has $10$ people, and for some mysterious reason their social network looks like this

Now suppose that Mr 1 sadly contracted corona virus ü§ßÔ∏è. But instead of putting him in isolation, the town decided to do social distancing more "fairly". Every day, among all the people who may still contract the virus through this network, one is chosen **uniformly at random** to put into quarantine. (Nothing is fairer than a lottery!)

## The records

Instead of giving everyone a lottery ticket each day, a more efficient way to do this is to randomly shuffle a deck of card of $1,.., 10$ and give each person a card.

The in day one, the person has card with the lowest number is chosen. On day two, the person among the remaining ones with a card of lowest number is chosen. This goes on until eventual until eventual Mr ü§ßÔ∏èÔ∏è is chosen.
